```python
train = pd.read_parquet('Data/train.parquet')
#try using log or bike count 
#One counter could be higher during the day and another one higher during the night
#manif
# 2 counter enter the dataset
train.drop(['counter_technical_id','counter_id','site_id','site_name'],axis=1,inplace=True)
train


# Convert 'date' column to datetime
train['date'] = pd.to_datetime(train['date'])

# Extract day, hour, month, year, and season from the 'date' column
train['day'] = train['date'].dt.day
train['hour'] = train['date'].dt.hour
train['month'] = train['date'].dt.month
train['year'] = train['date'].dt.year

# Define a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:  # 9, 10, 11
        return 'Autumn'

# Apply the function to get the season
train['season'] = train['month'].apply(get_season)

train

def get_week_end(nb_day) :
    if nb_day in list(range(0,4)):
        return 'Week_day'
    else :
        return 'Weekend'
    

    
train['Week_day']= train['date'].dt.dayofweek.apply(get_week_end)

max(train['date'])
```

```python
train = train.groupby(['counter_name','day','month','year']).apply(lambda x: x.sort_values('hour', ascending=True))
# Remove the names of the index levels
train.index.names = [None] * len(train.index.names)
# Now reset the index
train = train.reset_index(drop=True)

```

***Get the elevation :***

```python
def get_elevation(lat, lon):
    url = f"https://api.open-elevation.com/api/v1/lookup?locations={lat},{lon}"
    
    response = requests.get(url)
    if response.status_code == 200:
        elevation = response.json()['results'][0]['elevation']
        return elevation
    else:
        return "Error: Unable to get the elevation data"

elevation=[]
for i in range(0,len(train['latitude'].unique())):
    elevation.append(get_elevation(train['latitude'].unique()[i], train['longitude'].unique()[i])-35)
   

elevation_dict = dict(zip(train['latitude'].unique(), elevation))

# Map the elevation to the DataFrame based on Longitude
train['Elevation'] = train['latitude'].map(elevation_dict)
```

***Add confinement dates :***

```python
# Define your date range
conf_1_s = datetime(2020, 3, 17)
conf_1_e = datetime(2020, 5, 10)



conf_2_s = datetime(2020, 10, 30)
conf_2_e = datetime(2020, 11, 27)

conf_3_s=datetime(2021, 4, 3)
conf_3_e=datetime(2021, 5, 2)

conf_a1_s=datetime(2020, 5, 11)
conf_a1_e=datetime(2020, 6, 1)

conf_a2_s=datetime(2020, 11, 28)
conf_a2_e=datetime(2020, 12, 15)

c_v_1_s = datetime(2020, 12, 16)
c_v_1_e = datetime(2021, 4, 2)


c_v_2_s=datetime(2021, 5, 3)
c_v_2_e=datetime(2021, 6, 20)

#confinement
conf=[(datetime(2020, 3, 17),datetime(2020, 5, 10)),(datetime(2020, 10, 30),datetime(2020, 11, 27)),(conf_3_s,conf_3_e)]

#confinement_assouplie
conf_a=[(datetime(2020, 5, 11),datetime(2020, 6, 1)),(datetime(2020, 11, 28),datetime(2020, 12, 15))]

#couvre_feu
c_v=[(datetime(2020, 12, 16),datetime(2021, 4, 2)),(datetime(2021, 5, 3),datetime(2021, 6, 20))]

# Function to check if the date is in the range
def is_in_range(date):
    # Pour le confinement
    for start, end in conf : 
        if start <= date <= end :
            return 3
    for start, end in conf_a : 
        if start <= date <= end :
            return 2
    for start, end in c_v : 
        if start <= date <= end :
            return 1
    if date == datetime(2020, 12, 25):
        return 0
    return 0
    

# Apply the function to create a new column
train['Confinement'] = train['date'].apply(is_in_range)
train
```

***Split the adrress into direction and address:***

```python

# Split the address and direction into two columns
train[['Address', 'Direction']] = train['counter_name'].str.rsplit(' ', n=1, expand=True)

# Define a mapping from direction string to -1 or +1
direction_mapping = {
    'E-O': 1,
    'O-E': -1,
    'NO-SE' : 1,
    'SE-NO' : -1,
    'SO-NE' : 1 ,
    'SE-NO' : -1 ,
    'N-S' : 1,
    'S-N' : -1 ,
}

# Apply the mapping to the Direction column
train['Direction'] = train['Direction'].map(direction_mapping)

train
```
